```python
train=train.drop('Loan_ID',axis=1)
```

```python
train.head()
```

```python
train["Gender"]
train['Gender_cat'] = train['Gender'].map( {'Male':1, 'Female':0} )
train.drop('Gender', axis=1,inplace = True)
```

```python
train.columns
```

```python
train["Education"]
train['Education_cat'] = train['Education'].map( {'Graduate':1, 'Not Graduate':0,} )
train.drop('Education', axis=1,inplace = True)
```

```python
train['Married'].unique()
```

```python
train["Married"]
train['Married_cat'] = train['Married'].map( {'No':1, 'Yes':0, 'nan':2} )
train.drop('Married', axis=1,inplace = True)
```

```python
train['Property_Area'].unique()
```

```python
train["Property_Area"]
train['Property_Area_cat'] = train['Property_Area'].map( {'Urban':1, 'Rural':0,'Semiurban':2})
train.drop('Property_Area', axis=1,inplace = True)
```

```python
train.columns
```

```python
train
```

```python
train['Loan_Status'].unique()
```

```python
train["Loan_Status"]
train['Loan_Status_cat'] = train['Loan_Status'].map( {'Y':1, 'N':0,} )
train.drop('Loan_Status', axis=1,inplace = True)
```

```python
train
```

```python
train=train.drop('Self_Employed',axis=1)
```

```python
train
```

```python
X =train.drop('Loan_Status_cat',axis=1)
y = train['Loan_Status_cat']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
classifier = LogisticRegression()
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
# Accuracy score
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
classifier = LogisticRegression()
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
# Accuracy score
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,y_test))
```
